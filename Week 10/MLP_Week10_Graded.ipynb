```python
# prompt: Load the wine dataset from sklearn.

from sklearn.datasets import load_wine
X, y = load_wine(as_frame=True, return_X_y=True)
```

```python
X.head()
```

```python
y.head()
```

```python
# prompt: Split the dataset into train and test set with 70:30 ratio with random_state = 1

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)
```

```python
# prompt: Use AdaBoostClassifier with random_state = 1 (other values being default).

from sklearn.ensemble import AdaBoostClassifier
clf = AdaBoostClassifier(random_state=1)
clf.fit(X_train, y_train)
```

```python
# compute score on training data
print(clf.score(X_train, y_train))

# compute score on test data
print(clf.score(X_test, y_test))
```

```python
# prompt: Perform Hyperparameter tuning using GridSearchCV over the following:
# Use n_estimators as [100,500,1000]
# Use learning_rate as [0.5,1,2]
# Train the 'model' and compute the 'score' on training data and test data.

from sklearn.model_selection import GridSearchCV

param_dist = {
 'n_estimators': [100,500,1000],
 'learning_rate' : [0.5,1,2]
 }

grid_search = GridSearchCV(estimator=clf, param_grid=param_dist, cv=4, n_jobs=-1)

grid_search.fit(X_train, y_train)

# compute score on training data
print(grid_search.score(X_train, y_train))

# compute score on test data
print(grid_search.score(X_test, y_test))
```

```python
grid_search.best_params_
```

```python
# prompt: Write a code to predict the score on the training set and the testing set using Voting Classifier on wine dataset Write your code based on the following keypoints:
# Import model. Keep estimators as BaggingClassifier, Random Forest, GradientBoosting Classifier with random_state = 1

from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

# Define the individual classifiers
clf1 = BaggingClassifier(random_state=1)
clf2 = RandomForestClassifier(random_state=1)
clf3 = GradientBoostingClassifier(random_state=1)

# Create the VotingClassifier
eclf = VotingClassifier(
    estimators=[('bg', clf1), ('rf', clf2), ('gb', clf3)],
    voting='hard'  # Use majority voting
)

# Fit the VotingClassifier to the training data
eclf.fit(X_train, y_train)

# Compute score on training data
print(eclf.score(X_train, y_train))

# Compute score on test data
print(eclf.score(X_test, y_test))
```

```python
# prompt: used_cars.csv load

import pandas as pd

df2 = pd.read_csv('used_cars.csv')
```

```python
y=df2['MSRP']
```

```python
X=df2.drop('MSRP',axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
# prompt: Split the data into  train and test subsets using  70:30 ratio with `random_state = 1`.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# prompt: Train all the Estimators given below (with all default parameter values except it set random_state parameter equal to 1):
# BaggingRegressor(random_state=1)
# RandomForestRegressor(random_state=1)
# GradientBoostingRegressor(random_state=1)
# AdaBoostRegressor(random_state=1)

from sklearn.ensemble import BaggingRegressor, RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor

# Initialize the regressors
bagging_regressor = BaggingRegressor(random_state=1)
random_forest_regressor = RandomForestRegressor(random_state=1)
gradient_boosting_regressor = GradientBoostingRegressor(random_state=1)
ada_boost_regressor = AdaBoostRegressor(random_state=1)

# Fit the regressors to the training data
bagging_regressor.fit(X_train, y_train)
random_forest_regressor.fit(X_train, y_train)
gradient_boosting_regressor.fit(X_train, y_train)
ada_boost_regressor.fit(X_train, y_train)
```

```python
# prompt: find scores on test sets

# Evaluate the regressors on the test data
bagging_score = bagging_regressor.score(X_test, y_test)
random_forest_score = random_forest_regressor.score(X_test, y_test)
gradient_boosting_score = gradient_boosting_regressor.score(X_test, y_test)
ada_boost_score = ada_boost_regressor.score(X_test, y_test)

# Print the scores
print("Bagging Regressor Score:", bagging_score)
print("Random Forest Regressor Score:", random_forest_score)
print("Gradient Boosting Regressor Score:", gradient_boosting_score)
print("AdaBoost Regressor Score:", ada_boost_score)
```

```python
# prompt: On the same data set, perform the preprocessing stated in the previous questions, and perform Hyperparameter Tuning using GridSearchCV on AdaBoostRegressor. Use random_state to be 1.
# Hyperparameter tuning to be done over:
# Use n_estimators as [100,500,1000]
# Use learning_rate as [0.5,1,2]

param_dist = {
 'n_estimators': [100,500,1000],
 'learning_rate' : [0.5,1,2]
 }
clf = AdaBoostRegressor(random_state=1)
grid_search = GridSearchCV(estimator=clf, param_grid=param_dist, cv=4, n_jobs=-1)
grid_search.fit(X_train, y_train)
# compute score on training data
print(grid_search.score(X_train, y_train))
# compute score on test data
print(grid_search.score(X_test, y_test))
grid_search.best_params_
```
